## **Search for Target protein**

**Target search for coronavirus**

```python
#Target search for coronavirus
target = new_client.target
target_query = target.search('coronavirus')
targets = pd.DataFrame.from_dict(target_query)
targets
```

### Select and retrieve bioactivity for SARS coronavirus 3C-like porteinaise (6th entry)

We will assign the 7thth entry (which coressponds to the target protein, coronavirus 3C0like proteinaise to the __selected_table__

```python
selected_target = targets.target_chembl_id[6]
selected_target
```

Here, we will retrieve only bioactivity data for coronavirus-3C like proteinaise (CHEMBL3927) that are reported as IC50 values


```python
activity = new_client.activity
res = activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
```

```python
df = pd.DataFrame.from_dict(res)
```

```python
df
```

```python
df.head(3)
```

standard_value reflects the concentration of the drug: the lower the concentration required, the better

```python
df.to_csv('bioactivity_data.csv', index=False)
```

## Handling missing Data

If any compounds has the missing value for the __standard_value__ column then drop it

```python
df2 = df[df.standard_value.notna()]
df2
```

## Data pre-processing of the bioactivity data

### Labeling compounds are either being active, inactive, or intermediate

The bioactivity data is in the IC50 unit. Compounds having values of less than 1000 nM will be considered to be __active__ while those grater than 10, 000 nM will be considered to be __inactive__. As for those values in between 1000 and 10,000 nM wil be referred to as __intermediate__

```python
bioactivity_class = []
for i in df2.standard_value:
    if float(i) >= 10000:
        bioactivity_class.append("inactive")
    elif float(i) <= 1000:
        bioactivity_class.append("active")
    else:
        bioactivity_class.append("intermediate")
```

A compound is a molecule - which is a chemical structure that produces a modulatory activity (it exerts some effect on the target protein) in order to produce the desired biological effect (cures the symptoms).

Each molecule is identified with a *molecule_chembl_id*

If multiple rows share the same *molecule_chembl_id*, I will keep only one for simplicity

### Iterate the molecule_chembl_id to a list

```python
mol_cid = []
for i in df2.molecule_chembl_id:
    mol_cid.append(i)
```

### Iterate canonical_smiles to a list

SMILES are a way to represent molecules using ASCII strings

```python
canonical_smiles = []
for i in df2.canonical_smiles:
    canonical_smiles.append(i)
```

### Iterate standard_value to a list

```python
standard_value = []
for i in df2.standard_value:
    standard_value.append(i)
```

### Better Approach:

```python
selection = ['molecule_chembl_id', 'canonical_smiles', 'standard_value']
df3 = df2[selection]
```

```python
# add bioactivity class column with proper name
df3 = df2[['molecule_chembl_id', 'canonical_smiles', 'standard_value']].copy()
df3.loc[:, 'bioactivity_class'] = bioactivity_class
```

### Save dataframe to CSV file

```python
df3.to_csv('bioactivity_preprocessed_data.csv', index=False)
```
